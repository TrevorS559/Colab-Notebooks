# The Python Shell

Independent of Anaconda navigator, you can run Python commands directly from the command prompt or console of your operating system once you have installed it (via anaconda above, for example).  Open a command prompt and type "python" (without the quotes)

image.png


The triple > is the command prompt.  It tells you that the Python interpretor is waiting for a command.  We can run simple or complicated commands using the shell, but it is usually easier to work in the Spyder IDE because your work is saved and it is easier to work out what's going wrong with your code as you go along.  So we will be using the Spyder IDE mostly in this course.  In Spyder, the command prompt looks a little different. For Example:
